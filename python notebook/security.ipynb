## Visualization main points

1.**Are certain attack types more likely to come from specific attack sources?**

```python

plt.figure(figsize=(10, 5))
order = threats_df['Attack Type'].value_counts().index
sns.countplot(x='Attack Type', data=threats_df, order=order, palette='Blues_r')
plt.title("Frequency of Each Attack Type")
plt.xlabel("Attack Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()

```

the bar chart summarizes the frequency of each cyberattack type recorded in the dataset.

DDoS and Phishing pose the highest risk in terms of frequency, while Malware and Man-in-the-Middle attacks are relatively less common in this context. These insights can guide resource allocation and awareness training based on the most likely threats.

2.**Which countries or regions are the most common sources of attacks?**

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Country', data=threats_df,color='skyblue', order=threats_df['Country'].value_counts().index[:10])

order = threats_df['Country'].value_counts().index

sns.countplot(x='Country', data=threats_df, order=order, palette='Greens')
plt.title("Frequency of Each Country")
plt.xlabel("Country")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

The bar chart summarizes the number of cyberattacks recorded in each country within the dataset

The chart highlights that cyberattack frequency varies significantly across countries, with the "UK" being the most targeted and "China" the least in this dataset.

This may reflect differences in digital infrastructure, attack reporting, or attractiveness to cybercriminals.

3.**Which countries are at highest risk for specific attack types?**


```python
top_countries = threats_df.groupby('Country')['Financial Loss (in Million $)'].sum().nlargest(5).index
df_top = threats_df[threats_df['Country'].isin(top_countries)]


pivot_df = df_top.pivot_table(
    index='Country',
    columns='Attack Type',
    values='Financial Loss (in Million $)',
    aggfunc='sum',
    fill_value=0)
 
pivot_df.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')


plt.title("Top 5 Countries - Financial Losses by Attack Type", fontsize=18, fontweight='bold')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Total Financial Loss (in Million $)", fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

plt.legend(title='Attack Type', title_fontsize=12, fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The pivot bar chart summarizes the financial losses caused by different cyberattack types in the top 5 countries most affected by cyberattacks.

Australia and the UK experienced the highest financial losses due to DDoS attacks, indicating they were the most targeted by this attack type.

Brazil and Germany suffered the most financial impact from SQL Injection attacks, suggesting a common vulnerability in their systems.

Japan recorded the highest financial loss from Phishing attacks, highlighting user-targeted vulnerabilities as a major threat in that region.

These findings suggest the need for country-specific cybersecurity strategies that focus on the most relevant threat types.

4.**How does cyberattack impact (financial, users, resolution time) compare across different countries?**


```python
top_countries = threats_df.groupby('Country')['Number of Affected Users'].sum().nlargest(5).index
df_top = threats_df[threats_df['Country'].isin(top_countries)]


pivot_df = df_top.pivot_table(
    index='Country',
    columns='Attack Source',
    values='Number of Affected Users',
    aggfunc='sum',
    fill_value=0)
 
pivot_df.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='tab20')


plt.title("Top 5 Countries _ Affected Users by Attack Source", fontsize=18, fontweight='bold')
plt.xlabel("Country", fontsize=14)
plt.ylabel("Total Affected users", fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

plt.legend(title='Attack Source', title_fontsize=12, fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

The chart highlights that a large portion of cyberattacks remain unattributed, with “Unknown” sources being the most reported. This underlines the challenges in cyberattack forensics. Additionally, the presence of Nation-State attacks reflects geopolitical risks and the need for advanced defensive capabilities.

10.**What are the most exploited security vulnerabilities?**

```python
losses_by_vulnerability = threats_df.groupby('Security Vulnerability Type')['Financial Loss (in Million $)'].sum()
losses_by_vulnerability = losses_by_vulnerability[losses_by_vulnerability > 0]
plt.figure(figsize=(8, 8))
plt.pie(
    losses_by_vulnerability,
    labels=losses_by_vulnerability.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors)

plt.title(" Financial Loss by Security Vulnerability Type", fontsize=14)
plt.tight_layout()
plt.show()
```

The pie chart illustrates the percentage of financial losses associated with each Security Vulnerability Type.

Zero-day vulnerabilities accounted for the largest portion of financial loss, representing 26.1% of the total. This was closely followed by Social Engineering at 25.1%, and Unpatched Software at 24.4%.

These results highlight how critical it is to address emerging threats quickly (like zero-days), educate users to reduce social engineering risks, and maintain regular software updates to minimize exposure from known vulnerabilities.

11.**Does a particular industry face unique vulnerabilities compared to others?**

```python
industry_list = threats_df['Target Industry'].unique()

df_top = threats_df[threats_df['Target Industry'].isin(industry_list)]

pivot_df = df_top.pivot_table(
    index='Target Industry',
    columns='Security Vulnerability Type',
    values='Financial Loss (in Million $)',
    aggfunc='sum',
    fill_value=0)


pivot_df.plot(kind='bar', stacked=False, figsize=(10,6), colormap='Accent')

plt.title("Grouped Financial Losses in Industry by Vulnerability Type", fontsize=10)
plt.ylabel("Financial Loss (in Million $)")
plt.xlabel("Security Vulnerability Type")
plt.legend(title='Target Industry', bbox_to_anchor=(1.00, 1))
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()





```

In the IT industry, Zero-day vulnerabilities were responsible for the highest financial losses, highlighting the severe risk posed by unknown or unpatched exploits.

In the Education sector, Social engineering attacks led to the greatest share of losses, indicating the vulnerability of users to manipulation and phishing.

For both the Healthcare and Banking industries, the most significant financial impact came from Weak Password vulnerabilities, emphasizing the need for stronger authentication mechanisms in sensitive sectors.

These insights emphasize the importance of tailoring cybersecurity measures to the specific vulnerabilities most affecting each sector.

13.**Does a particular defense mechanism lead to faster incident resolution?**

```python
pivot_df = threats_df.pivot_table(
    index='Defense Mechanism Used',
    values='Incident Resolution Time (in Hours)',
    aggfunc='sum',
    fill_value=0)

pivot_df = pivot_df.T  
plt.figure(figsize=(12, 2))
sns.heatmap(pivot_df, annot=True, fmt='.0f', cmap='Greys', cbar=True)

plt.title("Total Incident Resolution Time by Defense Mechanism", fontsize=14)
plt.xlabel("Defense Mechanism Used")
plt.ylabel("")
plt.tight_layout()
plt.show()



pivot_df = threats_df.pivot_table(
    index='Defense Mechanism Used',
    columns='Attack Type',
    values='Incident Resolution Time (in Hours)',
    aggfunc='sum',
    fill_value=0
)

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_df, annot=True, fmt='.0f', cmap='YlGnBu', linewidths=0.5)

plt.title("Resolution Time by Defense Mechanism and Attack Type", fontsize=14)
plt.xlabel("Attack Type")
plt.ylabel("Defense Mechanism Used")
plt.tight_layout()
plt.show()
```

The heatmap chart displays the total resolution time taken by each defense mechanism (such as Antivirus, VPN, AI-based Detection) to mitigate different types of cyberattacks.

Antivirus and VPN show the longest resolution times overall, indicating that attacks take longer to resolve when using these methods.

Antivirus:
Takes the most time to mitigate SQL injection attacks.
Takes the least time in Man-in-the-Middle attacks.

VPN and AI-based Detection:
Show long resolution times for DDoS attacks.
Show the shortest resolution time for Malware and SQL Injection.

Some defense mechanisms are more efficient with specific types of attacks. This information can help organizations choose the most suitable defense tools for different threat types.

14.**Which defense mechanisms are most effective in reducing financial loss?**

```python
losses_by_defense = threats_df.groupby('Defense Mechanism Used')['Financial Loss (in Million $)'].sum()
plt.figure(figsize=(8, 8))
plt.pie(
    losses_by_defense,
    labels=losses_by_defense.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors)

plt.title(" Financial Loss by Defense Mechanism", fontsize=14)
plt.tight_layout()
plt.show()
```

The pie chart illustrates the percentage of financial losses associated with each cybersecurity defense mechanism.

Antivirus solutions accounted for the largest portion of the loss (21.4%), followed closely by VPN (20.3%) and Encryption (19.7%).

This distribution highlights that even common security tools may not fully prevent financial damage, and reinforces the need for multi-layered defense strategies and continuous evaluation of security effectiveness.
